```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel("SQL_Queries_Output.csv.xlsx")
df
```

```python
df.info
```

```python
category_stock = df.groupby('category')['stock'].sum()

plt.figure(figsize=(8,4))
category_stock.plot(kind='bar')
plt.title("Category-wise Total Stock")
plt.xlabel("Category")
plt.ylabel("Stock")
plt.grid(axis='y')
plt.show()
```

```python
category_price = df.groupby('category')['price'].mean()

plt.figure(figsize=(8,4))
category_price.plot(kind='bar')
plt.title("Average Price by Category")
plt.xlabel("Category")
plt.ylabel("Average Price")
plt.grid(axis='y')
plt.show()
```

```python
plt.figure(figsize=(8,4))
plt.hist(df['price'], bins=10)
plt.title("Price Distribution")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```
