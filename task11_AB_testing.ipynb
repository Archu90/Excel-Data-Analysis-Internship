```python
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
```

```python
data = {
    'group': ['A','A','A','A','A','A','B','B','B','B','B','B'],
    'room_rent': [8000, 8200, 8100, 8300, 8050, 8150,
                  9000, 9200, 9100, 9050, 9150, 9250]
}

df = pd.DataFrame(data)
df
```

```python
control = df[df['group'] == 'A']['room_rent']
test = df[df['group'] == 'B']['room_rent']
```

```python
control.mean(), test.mean()
```

```python
t_stat, p_value = stats.ttest_ind(control, test)
p_value
```

```python
if p_value < 0.05:
    decision = "Reject H0: Statistically significant difference"
else:
    decision = "Fail to Reject H0: No significant difference"

decision
```

```python
plt.boxplot([control, test], labels=['Control (A)', 'Test (B)'])
plt.title("Room Rent A/B Testing")
plt.ylabel("Room Rent")
plt.show()
```

```python
print("Control Mean:", control.mean())
print("Test Mean:", test.mean())
print("p-value:", p_value)
print(decision)
```
