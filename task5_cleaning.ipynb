```python
import pandas as pd
import numpy as np
df = pd.read_excel("SQL_Queries_Output.csv.xlsx")
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
for col in df.select_dtypes(include='number').columns:
  df[col].fillna(df[col].mean(), inplace=True)
for col in df.select_dtypes(include='object').columns:
  df[col].fillna(df[col].mode()[0], inplace=True)
```

```python
df.duplicated().sum()
```

```python
df['Record_Count'] = 1
```
